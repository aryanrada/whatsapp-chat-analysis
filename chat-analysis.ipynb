```python
import re
import pandas as pd
```

```python
a = open('WhatsApp Chat.txt', 'r', encoding='utf-8')
```

```python
chat = a.read()
print(chat)
```

```python
design = '\d{1,2}/\d{1,2}/\d{1,2},\s\d{1,2}:\d{1,2}\s/[a-zA-Z]+/g\s-\s'
```

```python
messages = re.split(design, chat)
print(messages)
```

```python
dates = re.findall(design, chat)
dates
```
